**Question 1.** What is hypothesis testing in statistics?
  -

Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It helps in deciding whether there is enough evidence in the sample data to support a certain belief (hypothesis) about a population parameter.

There are two primary hypotheses in any hypothesis test:

1. **Null Hypothesis (H₀):** This is the default or original claim that there is no effect or no difference. It is assumed true until evidence suggests otherwise.

2. **Alternative Hypothesis (H₁ or Ha):** This is the claim we are trying to find evidence for. It represents a new effect, difference, or relationship.

**Steps in Hypothesis Testing:**

1. Formulate H₀ and H₁.
2. Select a significance level (α), commonly 0.05.
3. Choose the appropriate test (e.g., z-test, t-test, ANOVA).
4. Compute the test statistic from sample data.
5. Determine the p-value or critical value.
6. Make a decision:
   - If p-value < α: Reject H₀.
   - If p-value ≥ α: Fail to reject H₀.

**Conclusion:** Hypothesis testing provides a formal procedure for comparing observed data with a claim (hypothesis), helping to make decisions with statistical backing.


**Question 2.**  What is the null hypothesis, and how does it differ from the alternative hypothesis?
  -

**Null Hypothesis (H₀):**
The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population. It is the assumption we test against.

  - **Alternative Hypothesis (H₁ or Ha):**
The alternative hypothesis is a statement that there is an effect, a difference, or a relationship. It represents what we aim to support.

**Key Difference:**
- H₀ assumes the status quo (no change).
- H₁ suggests a change or new effect.


**Question 3.** What is the significance level in hypothesis testing, and why is it important?
  -
**Significance Level (α):**
The significance level is a critical value in hypothesis testing that defines the probability of rejecting the null hypothesis (H₀) when it is actually true. This error is known as a Type I error.

**Why It's Important:**
The significance level determines how much risk we are willing to accept for making a wrong decision. For example, a significance level of 0.05 means there's a 5% chance of concluding that an effect exists when it really doesn't. It acts as a cutoff: if the p-value is less than α, we reject H₀.

Choosing the right α is crucial — a very small α reduces the risk of a Type I error but increases the risk of a Type II error (failing to reject a false H₀). Hence, α balances sensitivity and specificity in decision-making.



**Question 4.** What does a P-value represent in hypothesis testing?
  -
**P-value:**
The p-value is the probability of obtaining results at least as extreme as the observed results, assuming that the null hypothesis (H₀) is true.

**What It Represents:**
It measures the strength of evidence against the null hypothesis. A small p-value indicates that the observed data is unlikely under H₀, suggesting that H₀ may not be true.

**Interpretation:**
- If p-value ≤ α (e.g., 0.05): Reject H₀ (strong evidence against it).
- If p-value > α: Fail to reject H₀ (not enough evidence).

The p-value helps researchers determine whether an effect is statistically significant or likely due to random chance.


**Question 5.** How do you interpret the P-value in hypothesis testing?
  -
**Interpreting the P-value:** The p-value tells us how likely it is to observe our sample data (or something more extreme) if the null hypothesis (H₀) were true. It is a measure of evidence against H₀.

**General Interpretation Guidelines:**
- **p ≤ α (e.g., 0.05):** There is strong evidence against H₀, so we reject it. This means the observed result is statistically significant.
- **p > α:** There is not enough evidence to reject H₀, so we fail to reject it. This means the result is not statistically significant.

**Example:**
If p = 0.03 and α = 0.05, we reject H₀, suggesting the result is unlikely to occur by random chance alone.

The smaller the p-value, the stronger the evidence against the null hypothesis.


**Question 6.** What are Type 1 and Type 2 errors in hypothesis testing?
  -
**Type I Error (False Positive):**
Occurs when the null hypothesis (H₀) is true, but we incorrectly reject it. It means detecting an effect that doesn’t actually exist. The probability of making a Type I error is equal to the significance level α (e.g., 0.05).

  - **Type II Error (False Negative):**
Occurs when the null hypothesis is false, but we fail to reject it. It means missing an effect that actually exists. The probability of a Type II error is denoted by β, and (1 - β) is the power of the test.

**Summary:**
- Type I Error → Rejecting a true H₀.
- Type II Error → Failing to reject a false H₀.

Balancing these errors is important when designing statistical tests to ensure reliable and meaningful conclusions.



**Question 7.** What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  -
**One-Tailed Test:**
A one-tailed test checks for an effect in only one direction — either greater than or less than a certain value. It is used when the alternative hypothesis (H₁) specifies a direction, such as H₁: μ > μ₀ or H₁: μ < μ₀.

  - **Two-Tailed Test:**
A two-tailed test checks for an effect in both directions — it tests whether a parameter is simply different (not equal) from a specified value. The alternative hypothesis is of the form H₁: μ ≠ μ₀.

**Key Differences:**
- One-tailed test has more power to detect an effect in one direction but cannot detect an effect in the opposite direction.
- Two-tailed test is more conservative and checks for deviations in both directions, splitting the significance level α across two tails of the distribution.

**Example:**
- One-tailed: Testing if a drug *increases* recovery speed (H₁: μ > μ₀).
- Two-tailed: Testing if a drug *affects* recovery speed (H₁: μ ≠ μ₀), regardless of increase or decrease.



**Question 8.** What is the Z-test, and when is it used in hypothesis testing?
  -
**Z-test:**
The Z-test is a statistical test used to determine whether there is a significant difference between a sample mean and a known population mean, assuming the population standard deviation is known.

**When to Use a Z-test:**
- The sample size is large (typically n ≥ 30), or the population is normally distributed.
- The population standard deviation (σ) is known.
- You're testing hypotheses about population means or proportions.

**Common Applications:**
- One-sample Z-test: Testing if a sample mean differs from a known population mean.
- Two-sample Z-test: Comparing means from two independent large samples.
- Z-test for proportions: Comparing sample proportion to a population proportion.

The Z-test uses the standard normal distribution (Z-distribution) to calculate p-values and make decisions about the null hypothesis.



**Question 9.**  How do you calculate the Z-score, and what does it represent in hypothesis testing?
  -
**Z-score:**
The Z-score (or standard score) measures how many standard deviations a data point or sample mean is from the population mean under the null hypothesis.

**Formula:**
For a sample mean:
    Z = (X̄ - μ) / (σ / √n)
Where:
- X̄ = sample mean
- μ = population mean
- σ = population standard deviation
- n = sample size

**What It Represents:**
In hypothesis testing, the Z-score tells us how extreme the observed data is under the assumption that the null hypothesis is true. A large absolute Z-score suggests the sample mean is far from the population mean, which may indicate a statistically significant result.

The Z-score is then used to find the p-value by comparing it against the standard normal distribution.



**Question 10.** What is the T-distribution, and when should it be used instead of the normal distribution?
  -
**T-distribution:**
The t-distribution is a probability distribution used in hypothesis testing when the sample size is small and/or the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails, which allows for more variability.

**When to Use the T-distribution:**
- Sample size is small (typically n < 30).
- Population standard deviation (σ) is unknown.
- Data is approximately normally distributed.

**Why Use It Instead of the Normal Distribution:**
When σ is unknown, we estimate it using the sample standard deviation (s), which introduces extra uncertainty. The t-distribution adjusts for this uncertainty, especially in small samples, leading to more accurate confidence intervals and test results.

As the sample size increases, the t-distribution approaches the normal distribution.



**Question 11.** What is the difference between a Z-test and a T-test?
  -
**Z-test vs T-test:** Both Z-tests and T-tests are used to compare sample data to a population or between samples, but they differ based on sample size and known information.

**Z-test:**
- Used when the population standard deviation (σ) is known.
- Appropriate for large sample sizes (n ≥ 30).
- Based on the standard normal distribution.

**T-test:**
- Used when the population standard deviation is unknown.
- Suitable for small sample sizes (n < 30).
- Based on the t-distribution, which has heavier tails.

**Summary:**
- Use **Z-test** when σ is known and the sample is large.
- Use **T-test** when σ is unknown and/or the sample is small.

As sample size increases, the T-test results closely approximate the Z-test results.



**Question 12.** What is the T-test, and how is it used in hypothesis testing?
  -
**T-test:**
The t-test is a statistical test used to compare the means of one or two groups when the population standard deviation is unknown. It helps determine whether observed differences are statistically significant or likely due to random chance.

**Types of T-tests:**
1. **One-sample t-test** – Compares the sample mean to a known or hypothesized population mean.
2. **Independent two-sample t-test** – Compares means from two different independent groups.
3. **Paired sample t-test** – Compares means from the same group at different times (e.g., before and after a treatment).

**Usage in Hypothesis Testing:**
- Formulate H₀ (e.g., no difference in means) and H₁ (e.g., there is a difference).
- Calculate the t-statistic using sample data.
- Determine the degrees of freedom (df).
- Find the critical t-value or p-value.
- Make a decision to reject or fail to reject H₀.

The t-test is especially useful when dealing with small samples and unknown population variance.



**Question 13.** What is the relationship between Z-test and T-test in hypothesis testing?
  -
**Relationship Between Z-test and T-test:** The Z-test and T-test are closely related statistical methods used in hypothesis testing to assess whether a sample mean differs from a population mean or whether two sample means differ from each other.

**Key Relationship:**
- Both tests compare sample statistics to hypothesized population values.
- Both calculate a standardized test statistic (Z or t) using the difference between means and standard error.
- The main difference lies in the **assumptions about the population standard deviation** and **sample size**.

**When They Align:**
- As the sample size increases (typically n ≥ 30), the T-distribution approximates the normal distribution. So, the results of the T-test and Z-test become nearly identical for large samples.

**Summary:**
The T-test is a generalization of the Z-test that accounts for greater uncertainty when the population standard deviation is unknown. They follow similar logic and structure but are applied under different conditions.



**Question 14.** What is a confidence interval, and how is it used to interpret statistical results?
  -
**Confidence Interval (CI):**
A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean) with a certain level of confidence.

**Structure:**
CI = Estimate ± Margin of Error

For example, a 95% confidence interval means that if we repeated the sampling process many times, approximately 95% of the intervals calculated would contain the true population value.

**Usage in Interpretation:**
- Confidence intervals provide more information than a single point estimate by indicating the precision and uncertainty of the estimate.
- If a confidence interval for the difference between two means includes 0, it suggests that there may be no significant difference (we fail to reject H₀).
- In contrast, if the interval excludes 0, it indicates a statistically significant difference at the corresponding confidence level.

Confidence intervals are widely used in hypothesis testing and reporting results to show the reliability of estimates.



**Question 15.** What is the margin of error, and how does it affect the confidence interval?
  -
**Margin of Error (MoE):**
The margin of error is the amount added and subtracted from the sample estimate to create a confidence interval. It reflects the degree of uncertainty or variability in the estimate due to sampling.

**Formula (for mean with known σ):**
    MoE = Z * (σ / √n)
Where:
- Z is the Z-score based on the chosen confidence level (e.g., 1.96 for 95%)
- σ is the population standard deviation
- n is the sample size

**Effect on Confidence Interval:**
The margin of error determines the width of the confidence interval:
- A **larger margin of error** results in a **wider** confidence interval, indicating more uncertainty.
- A **smaller margin of error** leads to a **narrower** interval, indicating more precise estimates.

MoE is influenced by sample size (larger n reduces MoE), variability in data, and the chosen confidence level (higher confidence → higher MoE).



**Question 16.** How is Bayes' Theorem used in statistics, and what is its significance?
  -
**Bayes' Theorem:**
Bayes' Theorem is a fundamental concept in probability and statistics that describes how to update the probability of a hypothesis based on new evidence.

**Formula:**
P(A|B) = [P(B|A) * P(A)] / P(B)

Where:
- P(A|B) is the posterior probability (probability of A given B has occurred)
- P(B|A) is the likelihood (probability of B given A)
- P(A) is the prior probability of A
- P(B) is the total probability of B

**Usage in Statistics:**
Bayes' Theorem is used in statistical inference to revise predictions or beliefs in light of new data. It is especially valuable in:
- Medical diagnosis (updating disease probability based on test results)
- Spam filtering
- Machine learning (Naive Bayes classifiers)
- Decision-making under uncertainty

**Significance:**
It allows for a dynamic and evidence-based approach to probability, making it a powerful tool for probabilistic reasoning, especially when dealing with uncertainty or limited data.



**Question 17.** What is the Chi-square distribution, and when is it used?
  -
**Chi-square Distribution:**
The chi-square (χ²) distribution is a right-skewed probability distribution that arises when summing the squares of independent standard normal variables. It is defined only for non-negative values and depends on the degrees of freedom.

**When It Is Used:**
1. **Chi-square Test for Independence:** To determine whether two categorical variables are independent in a contingency table.
2. **Chi-square Goodness-of-Fit Test:** To assess whether a sample data distribution matches an expected distribution.
3. **Testing Variance:** To test hypotheses about population variance when the population is normally distributed.

**Key Features:**
- Used mainly for categorical data.
- The shape depends on the degrees of freedom (more degrees = more symmetrical).
- Critical for non-parametric hypothesis testing.

**Significance:**
The chi-square distribution enables statistical tests that help assess relationships between variables and evaluate how well observed data fit theoretical models.



**Question 18.** What is the Chi-square goodness of fit test, and how is it applied?
  -
**Chi-square Goodness-of-Fit Test:**
The chi-square goodness-of-fit test is used to determine whether observed categorical data matches a specific theoretical distribution or expected proportions.

**Purpose:**
It assesses whether the frequency distribution of observed data is consistent with what we would expect under a certain hypothesis.

**Formula:**
χ² = Σ [(Oᵢ - Eᵢ)² / Eᵢ]
Where:
- Oᵢ = observed frequency for category i
- Eᵢ = expected frequency for category i

**Steps to Apply:**
1. Define the null hypothesis (H₀: observed data follows the expected distribution).
2. Calculate expected frequencies based on the theoretical proportions.
3. Compute the chi-square statistic using the formula above.
4. Compare the test statistic with the critical value from the chi-square distribution table using appropriate degrees of freedom (df = number of categories - 1).
5. Make a decision: reject H₀ if the test statistic is greater than the critical value or if p-value < significance level.

**Use Case Example:**
Testing whether dice rolls are fair (equal probability for all sides).

This test helps validate assumptions about how well data fits a known distribution.



**Question 19.** What is the F-distribution, and when is it used in hypothesis testing?
  -
**F-distribution:**
The F-distribution is a continuous probability distribution used primarily to compare variances between two populations. It is right-skewed and depends on two degrees of freedom: one for the numerator and one for the denominator.

**Formula (F-statistic):**
F = (Variance₁ / Variance₂)

**When It Is Used:**
1. **ANOVA (Analysis of Variance):**
   - To test whether the means of three or more groups are significantly different.
   - It compares the variance **between** group means to the variance **within** the groups.

2. **Test for Equality of Variances:**
   - To check if two populations have equal variances before conducting t-tests.

**Key Features:**
- Always positive (since variances are positive).
- The shape of the distribution changes with degrees of freedom.
- Sensitive to sample size and variance differences.

**Significance:**
The F-distribution is crucial for tests involving multiple samples and helps determine if differences among group means or variances are statistically significant.



**Question 20.** What is an ANOVA test, and what are its assumptions?
  -
**ANOVA (Analysis of Variance):**
ANOVA is a statistical method used to compare the means of three or more independent groups to determine if at least one group mean is significantly different from the others.

**How It Works:**
It analyzes the ratio of:
- **Between-group variance** (variation due to the treatment or group differences)
- **Within-group variance** (natural variation within each group)

If the between-group variance is much larger than the within-group variance, the F-statistic will be large, suggesting significant differences.

**Null Hypothesis (H₀):** All group means are equal.
**Alternative Hypothesis (H₁):** At least one group mean is different.

**Assumptions of ANOVA:**
1. **Independence** – Observations must be independent across and within groups.
2. **Normality** – Data in each group should be approximately normally distributed.
3. **Homogeneity of variances** – All groups should have similar variances.

ANOVA is commonly used in experiments, surveys, and observational studies to test the effect of categorical factors on continuous outcomes.



**Question 21.** What are the different types of ANOVA tests?
  -
**Types of ANOVA Tests:**
ANOVA comes in different forms depending on the number of factors (independent variables) and the design of the study.

1. **One-Way ANOVA:**
   - Used when comparing the means of **three or more groups** based on **one independent variable**.
   - Example: Comparing test scores across three teaching methods.

2. **Two-Way ANOVA:**
   - Used to assess the effect of **two independent variables** on a dependent variable.
   - Can also detect **interaction effects** between the two factors.
   - Example: Studying the effect of teaching method and gender on test scores.

3. **Repeated Measures ANOVA:**
   - Used when the **same subjects** are measured multiple times (within-subject design).
   - Example: Measuring blood pressure of patients before, during, and after treatment.

4. **MANOVA (Multivariate ANOVA):**
   - Extension of ANOVA when there are **multiple dependent variables**.
   - Tests for the effect of one or more independent variables on several related outcomes.

Each type of ANOVA helps in analyzing complex experimental designs and understanding the influence of different factors.



**Question 22.** What is the F-test, and how does it relate to hypothesis testing?
  -
**F-test:**
The F-test is a statistical test that compares two variances to determine if they are significantly different. It is based on the F-distribution and is commonly used to test hypotheses involving variability or multiple group means.

**Formula:**
F = (Variance₁ / Variance₂)
Where Variance₁ is usually the larger sample variance.

**Relation to Hypothesis Testing:**
- In hypothesis testing, the F-test evaluates whether observed differences in variances or group means could have occurred by random chance.
- It's most commonly used in **ANOVA**, where the F-test checks if the variability between group means is significantly greater than the variability within groups.

**Hypotheses in ANOVA F-test:**
- H₀: All group means are equal.
- H₁: At least one group mean is different.

If the computed F-statistic is greater than the critical value (based on the F-distribution and degrees of freedom), we reject H₀, suggesting a significant difference exists.

**Significance:**
The F-test helps assess models and compare groups, playing a key role in experimental design and regression analysis.

**Question 3.**  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
# Function to Calculate the PDF of a Bernoulli Distribution

def bernoulli_pdf(x, p):
    """
    Calculates the PDF of a Bernoulli distribution.

    Parameters:
    x (int): Value (0 or 1)
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
      raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
      raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.6))
print("P(X=0):", bernoulli_pdf(0, 0.6))
```

**Question 9.** Write a Python function to calculate Z-scores from a dataset and plot them.

```python
# Function to Calculate Z-scores and Plot Them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores from the dataset and plots them.

    Parameters:
    data (list or np.array): Input numerical data

    Returns:
    np.array: Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.plot(z_scores, marker='o', linestyle='none')
    plt.axhline(y=0, color='r', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# Example usage
sample_data = np.random.normal(50, 10, 100)
z_scores = calculate_and_plot_z_scores(sample_data)
```

**Question 12.** Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).

**Question 5.** Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
# Visualizing Type I and Type II Errors in Hypothesis Testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualizes Type I and Type II errors for hypothesis testing.

    Parameters:
    - mu0: Null hypothesis mean
    - mu1: Alternative hypothesis mean
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value under H0
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # Plotting range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions under H0 and H1
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(x, y0, label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H₁: μ = {}'.format(mu1), color='green')

    # Shade Type I Error Region (α)
    plt.fill_between(x, y0, where=(x > x_crit), color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II Error Region (β)
    plt.fill_between(x, y1, where=(x < x_crit), color='orange', alpha=0.3, label='Type II Error (β)')

    plt.axvline(x_crit, color='black', linestyle='--', label='Critical Value')
    plt.title('Type I and Type II Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculate Type II Error (β)
    beta = norm.cdf(x_crit, loc=mu1, scale=se)
    print(f"Critical Value (x_crit): {x_crit:.2f}")
    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Call the function
visualize_type1_type2_errors()
```

**Question 9.** Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
# Function to calculate confidence interval for a sample mean

import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a sample.

    Parameters:
    - sample: list or NumPy array of sample data
    - confidence: confidence level (default is 95%)

    Returns:
    - (lower_bound, upper_bound): tuple of confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error
    return lower, upper

# Example usage:
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=25)
ci_lower, ci_upper = confidence_interval(sample_data)

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Python program to calculate the margin of error for a given confidence level

import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
    - sample: list or NumPy array of sample values
    - confidence: desired confidence level (default is 0.95 for 95%)

    Returns:
    - margin_of_error: the margin of error value
    """
    n = len(sample)
    std_dev = np.std(sample, ddof=1)
    se = std_dev / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_crit * se
    return moe

# Example usage:
np.random.seed(1)
data = np.random.normal(loc=60, scale=12, size=40)
moe = margin_of_error(data)

print(f"Sample Mean: {np.mean(data):.2f}")
print(f"Margin of Error (95% confidence): ±{moe:.2f}")
```

**Question 11.** Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
# Bayesian Inference using Bayes' Theorem in Python

def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    """
    Apply Bayes' Theorem to compute the posterior probability P(A|B).

    Parameters:
    - prior_A: P(A) - Prior probability of hypothesis A
    - likelihood_B_given_A: P(B|A) - Likelihood of evidence B given A
    - prior_not_A: P(~A) - Prior probability of not A
    - likelihood_B_given_not_A: P(B|~A) - Likelihood of evidence B given not A

    Returns:
    - posterior: P(A|B) - Posterior probability
    """
    numerator = likelihood_B_given_A * prior_A
    denominator = numerator + (likelihood_B_given_not_A * prior_not_A)
    posterior = numerator / denominator
    return posterior

# Example scenario:
# Suppose a disease affects 1% of the population (P(Disease) = 0.01)
# The test detects the disease correctly 99% of the time (P(Positive|Disease) = 0.99)
# But it also gives a false positive 5% of the time (P(Positive|No Disease) = 0.05)

prior = 0.01
likelihood_given_disease = 0.99
prior_not = 1 - prior
likelihood_given_no_disease = 0.05

posterior = bayes_theorem(prior, likelihood_given_disease, prior_not, likelihood_given_no_disease)

print(f"Posterior probability of having the disease given a positive test: {posterior:.4f}")
```

**Question 12.** Perform a Chi-square test for independence between two categorical variables in Python.


```python
# Chi-square test for independence between two categorical variables

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table (observed frequencies)
# Example: Gender vs Preference
data = {
    'Likes_A': [20, 30],  # Male, Female
    'Likes_B': [25, 25]
}
# Rows: Gender (Male, Female), Columns: Product preference (A, B)
contingency_table = pd.DataFrame(data, index=['Male', 'Female'])

# Step 2: Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Step 3: Print results
print("Contingency Table:")
print(contingency_table)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))
print(f"\nChi-square Statistic: {chi2:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: The variables are dependent.")
else:
    print("Fail to reject the null hypothesis: The variables are independent.")
```

**Question 14.** Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
# Chi-square Goodness-of-Fit Test using Python

import numpy as np
from scipy.stats import chisquare

# Step 1: Define observed and expected frequencies
observed = np.array([18, 22, 20, 40])   # Example observed counts
expected = np.array([25, 25, 25, 25])   # Expected under uniform distribution

# Step 2: Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 3: Print results
print("Observed frequencies:", observed)
print("Expected frequencies:", expected)
print(f"\nChi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")
```
